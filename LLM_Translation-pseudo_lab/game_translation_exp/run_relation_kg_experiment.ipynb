```python
import os
import sys
import subprocess
from pathlib import Path
import pandas as pd

try:
    from dotenv import load_dotenv
except ImportError:
    load_dotenv = None


BASE_OVERRIDE = os.getenv('GAME_TRANSLATION_BASE_DIR', '').strip()
if BASE_OVERRIDE:
    BASE = Path(BASE_OVERRIDE).expanduser().resolve()
else:
    cwd = Path.cwd().resolve()
    candidates = [cwd, cwd / 'game_translation_exp', cwd.parent / 'game_translation_exp']
    BASE = next((p for p in candidates if (p / 'data').exists() and (p / 'scripts').exists()), cwd)

if not (BASE / 'data').exists():
    raise FileNotFoundError(f'Could not locate game_translation_exp base dir from cwd={Path.cwd()}')
SCRIPTS = BASE / 'scripts'
DATA = BASE / 'data'
REL = DATA / 'relation_kg'
REL_EXT = DATA / 'relation_kg_external'
OUT = BASE / 'outputs'
EVAL = BASE / 'eval'

if load_dotenv is not None:
    load_dotenv(BASE.parent / '.env', override=True)

RUN_DATE = os.getenv('RUN_DATE', '2026-04-12')
DRY_RUN = os.getenv('DRY_RUN', 'true').lower() == 'true'

print('RUN_DATE =', RUN_DATE)
print('DRY_RUN =', DRY_RUN)
print('OPENAI_API_KEY loaded =', bool(os.getenv('OPENAI_API_KEY')))
```

```python
# 1) Optional: refresh auto candidates (not canonical)
REFRESH_AUTO = False
if REFRESH_AUTO:
    cmd = [sys.executable, str(SCRIPTS / 'extract_relation_candidates.py'), '--allow-cooccurrence']
    print('Running:', ' '.join(cmd))
    subprocess.run(cmd, check=True)
```

```python
# 2) Build sample relation context from external curated edges
cmd = [
    sys.executable,
    str(SCRIPTS / 'build_relation_context_external.py'),
    '--samples-csv', str(DATA / 'samples.csv'),
    '--edges-csv', str(REL_EXT / 'relation_edges_external_v1.csv'),
    '--alias-csv', str(REL_EXT / 'entity_aliases_external.csv'),
    '--out-csv', str(REL_EXT / 'sample_relation_context_external.csv'),
]
print('Running:', ' '.join(cmd))
subprocess.run(cmd, check=True)
```

```python
# 3) Inspect rows with explicit external relation context
ctx = pd.read_csv(REL_EXT / 'sample_relation_context_external.csv')
has_ctx = ctx[~ctx['relation_context'].str.contains('No external relation context', na=False)]
print('with context:', len(has_ctx), '/', len(ctx))
ctx[['sample_id','detected_entities','relation_context']].head(10)
```

```python
# 4) Run E condition (external KG)
cmd = [sys.executable, str(SCRIPTS / 'run_condition_e_external.py'), '--run-date', RUN_DATE]
if DRY_RUN:
    cmd.append('--dry-run')
print('Running:', ' '.join(cmd))
subprocess.run(cmd, check=True)

df_e = pd.read_csv(OUT / f'run_{RUN_DATE}_external' / 'E_outputs_external.csv')
print('E outputs:', len(df_e))
df_e.head(10)
```

```python
# 5) Build A/C/E comparison sheet
run_dir = OUT / f'run_{RUN_DATE}'
run_ext_dir = OUT / f'run_{RUN_DATE}_external'
samples = pd.read_csv(DATA / 'samples.csv')

A = pd.read_csv(run_dir / 'A_outputs.csv') if (run_dir / 'A_outputs.csv').exists() else None
C = pd.read_csv(run_dir / 'C_outputs.csv') if (run_dir / 'C_outputs.csv').exists() else None
E = pd.read_csv(run_ext_dir / 'E_outputs_external.csv') if (run_ext_dir / 'E_outputs_external.csv').exists() else None

merged = samples[['sample_id','sentence_type','source_text']].copy()
if A is not None:
    merged = merged.merge(A[['sample_id','translation']].rename(columns={'translation':'condition_A'}), on='sample_id', how='left')
if C is not None:
    merged = merged.merge(C[['sample_id','translation']].rename(columns={'translation':'condition_C'}), on='sample_id', how='left')
if E is not None:
    merged = merged.merge(E[['sample_id','translation']].rename(columns={'translation':'condition_E'}), on='sample_id', how='left')

for col in [
    'A_meaning','A_term','A_consistency','A_naturalness','A_ui_fit',
    'C_meaning','C_term','C_consistency','C_naturalness','C_ui_fit',
    'E_meaning','E_term','E_consistency','E_naturalness','E_ui_fit',
    'best','error_type','notes'
]:
    if col not in merged.columns:
        merged[col] = ''

out_eval = EVAL / f'eval_sheet_prefilled_ACE_{RUN_DATE}.csv'
merged.to_csv(out_eval, index=False, encoding='utf-8')
print('saved:', out_eval)
print('rows:', len(merged))
```
